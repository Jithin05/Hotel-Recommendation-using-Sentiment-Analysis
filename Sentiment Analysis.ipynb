# TEXT PREPROCESSING 
### DELETING NaN ROWS

```python
df1=df.dropna(axis=0)
df1
```

### DELETING DUPLICATE ROWS

```python
df_drop = df1.drop_duplicates()
df_drop
```

### CLEANING ROWS WITH NON-ENGLISH REVIEWS

```python
df_drop = df_drop[df_drop['Review'].map(lambda x: x.isascii())] 
df_drop
```

```python
def remove_numbers(text):
    pattern=r'[^A-z\s]'
    return re.sub(pattern, '',text).lower()
df_drop['Review']=df_drop.Review.apply(remove_numbers)
df_drop
```

### LEMMATIZATION ,TOKENIZATION AND REMOVAL OF STOP WORDS

```python
nltk.download('all')
```

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
df_drop['text_lemmatized']=df_drop.Review.apply(lemmatize_text)
stop_words=stopwords.words('english')
df_drop['cleaned_review']=df_drop['text_lemmatized'].apply(lambda x:[item for item in x if item not in stop_words])
df_drop
```

```python
df_reviews=df_drop[["Hotel_name","cleaned_review"]]
df_reviews["cleaned_review"]=df_reviews.cleaned_review.apply(lambda x:' '.join(x))
df_reviews
```

```python
group_by= df_reviews['cleaned_review'].groupby(df_reviews['Hotel_name']).sum()
df_grouped=pd.DataFrame(group_by).reset_index()
df_grouped
```

# SENTIMENTAL ANALYSIS


```python
def SentimentAnalyzer(text):
    sid_obj= SentimentIntensityAnalyzer()
    return(sid_obj.polarity_scores(text)) 
df_reviews["polarity_scores"]=df_reviews.cleaned_review.apply(SentimentAnalyzer)
df_reviews
```

```python
def get(dict):
    return(dict.get("compound"))
def analysis(score):
    if score==0:
        return("neutral")
    elif score>=0:
        return("positive")
    else:
        return("negative")
    
df_reviews["compound"]=df_reviews.polarity_scores.apply(get)
df_reviews["emotion"]=df_reviews.compound.apply(analysis)
df_reviews
```

```python
grouped_emotion=df_reviews[["Hotel_name","emotion"]]
grouped_emotion
```

```python
grouped_compound=df_reviews["compound"].groupby(df_reviews["Hotel_name"]).mean()
grouped_compound=pd.DataFrame(grouped_compound).reset_index()
grouped_compound=grouped_compound.sort_values(by="compound")
grouped_compound
```

```python
s=pd.DataFrame(grouped_emotion.groupby('Hotel_name')['emotion'].value_counts("count"))
s.rename(columns={"emotion":"weighted_average"})
```
